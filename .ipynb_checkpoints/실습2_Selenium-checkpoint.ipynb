# 비트코인 데이터 수집하기

https://coinmarketcap.com

### 사이트 접속하기

```python
from selenium import webdriver
from bs4 import BeautifulSoup

driver = webdriver.Chrome('C:/chromedriver')
# driver = webdriver.Chrome('~/Downloads/chromedriver')
driver.get('https://coinmarketcap.com/')
```

### 코인 상세페이지 URL 가져오기

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
links = soup.select('tr span a.link-secondary')
urls = []
for link in links:
    urls.append(link.attrs["href"])
```

### URL 접근하기

```python
driver.get('https://coinmarketcap.com/' + urls[0] +"historical-data/")
```

### 데이터 수집하기

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
rows = soup.select('tr.text-right')
stock_list = []
for row in rows:
    cols = row .select('td')
    dict = {}
    for idx, col in enumerate(cols):
        if idx==0:
            dict["날짜"] = col.text
        elif idx==1:
            dict["시가"] = col.text
        elif idx==2:
            dict["고가"] = col.text
        elif idx==3:
            dict["저가"] = col.text
        elif idx==4:
            dict["종가"] = col.text
        elif idx==5:
            dict["거래량"] = col.text
        elif idx==6:
            dict["시가총액"] = col.text
    stock_list.append(dict)
```

## 함수화 시켜보기

### 해당 URL에 있는 주식 값들 가져오는 함수

```python
def getStock(url):
    driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    rows = soup.select('tr.text-right')
    stock_list = []
    for row in rows:
        cols = row .select('td')
        dict = {}
        for idx, col in enumerate(cols):
            if idx==0:
                dict["날짜"] = col.text
            elif idx==1:
                dict["시가"] = col.text
            elif idx==2:
                dict["고가"] = col.text
            elif idx==3:
                dict["저가"] = col.text
            elif idx==4:
                dict["종가"] = col.text
            elif idx==5:
                dict["거래량"] = col.text
            elif idx==6:
                dict["시가총액"] = col.text
        stock_list.append(dict)
    return stock_list
```

```python
getStock('https://coinmarketcap.com/' + urls[0] +"historical-data/")
```
