# Selenium 실습

```python
from selenium import webdriver

# 아까 받은 chromedriver의 위치를 지정해준다.
driver = webdriver.Chrome('./chromedriver')
driver.get('https://nid.naver.com/nidlogin.login')
```

## 태그의 속성 값(name) 사용해 입력하기

```python
driver.find_element_by_name('id').send_keys('naver_id')
driver.find_element_by_name('pw').send_keys('mypassword1234')
```

## 태그의 ID를 통해 값 입력하기

```python
driver.find_element_by_id('id').send_keys('naver_id')
driver.find_element_by_id('pw').send_keys('mypassword1234')
```

## 로그인 버튼 클릭하기

```python
driver.find_element_by_class_name('btn_global').click()
# driver.find_element_by_xpath('//*[@id="frmNIDLogin"]/fieldset/input').click()
```

## 검색어에 따른 네이버 블로그 타이틀 크롤링
### 네이버에 접속

```python
from selenium import webdriver

# 아까 받은 chromedriver의 위치를 지정해준다.
driver = webdriver.Chrome('./chromedriver')
driver.get('https://www.naver.com')
```

### 검색어 입력!

```python
keyword = "치앙마이"
driver.find_element_by_id('query').send_keys(keyword)
```

### 검색 버튼 클릭

```python
driver.find_element_by_id('search_btn').click()
```

### 블로그 메뉴 클릭

```python
driver.find_element_by_xpath('//*[@id="lnb"]//li[@class="lnb3"]').click()
```

### 웹 소스 가져오기

```python
html = driver.page_source
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')
soup
```

### 블로그 타이틀 추출하기

```python
blogs = soup.select('a.sh_blog_title')
for blog in blogs:
    print(blog.text)
```
